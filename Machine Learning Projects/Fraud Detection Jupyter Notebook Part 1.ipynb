```python
#Create function which splits the data via random sampling

import numpy as np

def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data)*test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
#Create a function which randomly samples by id (will be created later).  To ensure the same test set is generated each time, we
#compute a hash of each instance's identifier, keep the last byte of the hash and put the instance in the test set if this value 
#is lower or equal to 51.
#So, the new test set will contain 20% of the new instances, but will not contain any instance previously in the training set.

import hashlib

def test_set_check(identifier, test_ratio, hash):
    return hash(np.int64(identifier)).digest()[-1] < 256*test_ratio

#random sampling

def split_train_test_by_id(data, test_ratio, id_column, hash=hashlib.md5):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio, hash))
    return data.loc[~in_test_set], data[in_test_set]

```

```python
#Split the data 70/30 for training/validation
credit_with_Primary_Key = credit.reset_index()
train_set, test_set = split_train_test_by_id(credit_with_Primary_Key, 0.3, "index")
```

```python
#Prepare data for machine learning
fraud = train_set.drop("Class", axis=1)
fraud_labels = train_set["Class"].copy()
```

```python
#Prepare data for machine learning
fraud_clean = fraud.drop("index", axis=1)
```

```python
#Run logistic regression algo for binary classification

from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(max_iter=1000, solver='lbfgs')
log_reg.fit(fraud_clean, fraud_labels)
```

```python
#Run the computed model on new data and determine model performance via confusion matrix and auc score

from sklearn.metrics import confusion_matrix

test_actual = test_set["Class"].copy()
test_clean1 = test_set.drop("Class", axis=1)
test_clean = test_clean1.drop("index", axis=1)

test_predicted = log_reg.predict(test_clean)
confusion_matrix(test_actual, test_predicted)
```

```python
#Decent AUC score and good accuracy score of 99.9%, however, the fraud rate is ~0.2%, which is unfortunate, since we have a 33% false negative rate.
#Next step(s): 1) try stratified sampling (to match fraud distribution) and then logistic regression
#              2) try random forest binary decision tree classification (relationship between independent variables and dependent variable likely non-linear) CAREFUL - THIS CAN LEAD TO OVERFITTING
#              3) try Amazon Machine Learning's Binary Classification Suite
```

```python
#Split data via stratified sampling
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.3, random_state=42)
for train_index, test_index in split.split(credit, credit["Class"]):
    strat_train_set = credit.loc[train_index]
    strat_test_set = credit.loc[test_index]
```

```python
strat_fraud_clean = strat_train_set.drop("Class", axis=1)
strat_fraud_labels = strat_train_set["Class"].copy()
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(max_iter=1000, solver='lbfgs')
log_reg.fit(strat_fraud_clean, strat_fraud_labels)
```

```python

strat_test_actual = strat_test_set["Class"].copy()
strat_test_clean = strat_test_set.drop("Class", axis=1)
strat_test_predicted = log_reg.predict(strat_test_clean)
confusion_matrix(strat_test_actual, strat_test_predicted)
```
